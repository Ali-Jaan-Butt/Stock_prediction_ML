<h1>Cleaning Data from dataset</h1>

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.info() # Data information
```

```python
df.sample(5) # Sample unclean data
```

```python
# Cleaning columns
def kms(k):
    return k.replace(' kms','')
def kc(c):
    return c.replace(',','')
df['kms_driven'] = df['kms_driven'].apply(kms)
df['kms_driven'] = df['kms_driven'].apply(kc)
```

```python
# Cleaning columns
def price(p):
    return p.replace(',','')
df['Price'] = df['Price'].apply(price)
df['Price'] = df['Price'].replace('Ask For Price',None)
```

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
# Drop null rows
df = df.dropna()
df = df.reset_index()
del df['index']
```

```python
# Changing data types of the required columns to required datatypes
df['year'] = df['year'].astype(int)
df['Price'] = df['Price'].astype(int)
df['kms_driven'] = df['kms_driven'].astype(float)
```

```python
# Saving cleaned data to file
df.to_csv('clean_data.csv',index=False)
```
